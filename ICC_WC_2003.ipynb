```python
playerList= ['Tendulkar','Ganguly','Ponting','Gilchrist','Gibbs','Atapattu','Andy','Hayden','Symonds','Martyn']
playerArray=np.array(playerList)
playerArray
```

```python
print playerArray[0:6]
print playerArray[:7]
print playerArray[3:]
print playerArray[3:8]
```

```python
runList=[673,465,415,408,384,382,332,328,326,323]
runArray=np.array(runList)
runArray
```

```python
matchList=[11,11,10,10,6,10,7,11,5,8]
matchArray=np.array(matchList)
matchArray
```

```python
avgList=[]
for x in range(len(matchArray)):
    avgList.append(runArray[x]/matchArray[x])
avgArray=np.array(avgList)
avgArray
    
```

```python
dic={'player':playerArray,'match':matchArray,'runs':runArray,'avg':avgArray}
print dic
```

```python
scoreCard=pd.DataFrame(dic, columns = ['player','match','runs','avg'])
scoreCard
```

```python
scoreCard.describe()
```

```python
sortedScore=scoreCard.sort_values(by=['runs'],ascending= False)
sortedScore.set_index("player",inplace = True)
sortedScore.plot.hist(alpha=0.7)
plt.show()
sortedScore.plot()
plt.show()
```

```python
explode = [0.5,0.3,0.1,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
fig1, ax1 = plt.subplots()
ax1.pie(scoreCard['runs'], explode=explode, labels=scoreCard['player'], autopct='%1.0f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
